```python
#  Creation of new customer accounts with unique account numbers and initial balances (creation of data entities and attributes).
# into a Dictionary
```

```python
current_account_number = 1000
all_accounts = {}

def create_account():
    global current_account_number

    cust_name = input("Enter your full name: ")
    cust_phone = input("Enter your phone number: ")
    cust_dob = input("Enter your date of birth (DD-MM-YYYY): ")
    cust_age = input("Enter your age: ")
    cust_gender = input("Are you male or female: ")
    cust_email = input("Enter your email: ")
    
    cust_pin = int(input("Create a 4-digit PIN for your account: "))

    current_account_number += 1
    account_num = current_account_number

    account_details = {
        "Account_num": account_num,
        "PIN": cust_pin,  
        "Balance": 0,
        "Cust_name": cust_name,
        "Phone_num": cust_phone,
        "Dob": cust_dob,
        "Gender": cust_gender,
        "Age": cust_age,
        "Email": cust_email,
        "History": []
    }

    all_accounts[account_num] = account_details

    print("\nAccount created successfully!")
    print("Your account number is:", account_num)
    print("Please remember your PIN:", cust_pin)

create_account()
```

```python
# Implemention of a Deposit Functionality (in a nested dictionary where nested functionality is being used (if)
```

```python
all_accounts = {
    1001: {
        "Account_num": 1001,
        "Balance": 100,
        "Cust_name": "Test User",
        "Phone_num": "0000000000",
        "History": []
    }
}

def do_deposit():
    acct_num = int(input("What is your account number?: "))
    deposit_amt = int(input("How much do you want to deposit?: "))

    if acct_num in all_accounts:
        all_accounts[acct_num]["Balance"] += deposit_amt
        all_accounts[acct_num]["History"].append(f"Deposited {deposit_amt}")
        print("Deposit successful!")
        print("New balance:", all_accounts[acct_num]["Balance"])
    else:
        print("Account not found.")

do_deposit()
```

```python
# Implemention of a Withdrawal Functionality (in a nested dictionary where nested functionality is being used (if)
```

```python

all_accounts = {
    1001: {
        "Account_num": 1001,
        "Balance": 600,
        "Cust_name": "Test User",
        "Phone_num": "0000000000",
        "History": []
    }
}

def do_withdrawal():
    acct_num = int(input("What is your account number?: "))
    withdraw_amt = int(input("How much do you want to withdraw?: "))

    if acct_num in all_accounts:
        if all_accounts[acct_num]["Balance"] >= withdraw_amt:
            all_accounts[acct_num]["Balance"] -= withdraw_amt
            all_accounts[acct_num]["History"].append(f"Withdrew {withdraw_amt}")
            print("Withdrawal successful!")
            print("New balance:", all_accounts[acct_num]["Balance"])
        else:
            print("Insufficient balance.")
    else:
        print("Account not found.")

do_withdrawal()
```

```python
# Implemention of a Balance Inquiry using the def functionality in order to define the function check _balance 
# making if to make decisions within the function
```

```python
def check_balance():
    acct_num = int(input("What is your account number?: "))

    if acct_num in all_accounts:
        print("Your current balance is:", all_accounts[acct_num]["Balance"])
    else:
        print("Account not found.")

check_balance()
```

```python
# History of transactions
```

```python
all_accounts = {
    1001: {
        "Account_num": 1001,
        "Cust_name": "Ifeoma",
        "Phone_num": "0000000000",
        "Transactions": [("Deposit", 600), ("Withdraw", 200)]
    }
}

def calculate_balance(transactions):
    balance = 0
    for action, amount in transactions:
        if action == "Deposit":
            balance += amount
        elif action == "Withdraw":
            balance -= amount
    return balance

def view_transaction_history():
    acct_num = int(input("Enter your account number: "))

    if acct_num in all_accounts:
        transactions = all_accounts[acct_num]["Transactions"]

        
        balance = calculate_balance(transactions)

        print("\n--- Transaction History ---")
        for action, amount in transactions:
            print(f"{action}ed {amount}")
        print("Current Balance:", balance)
    else:
        print("\n--- Transaction History ---")
        print("Account not found.")


view_transaction_history()


```

```python
# Secure Save end menu
```

```python
def run_Securesave_app():
    while True:
        print("\n=== Secure Save menu ===")
        print("1. Check Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. View Transaction History")
        print("5. Exit")

        choice = input("Choose an option (1-5): ")

        if choice == "1":
            check_balance()
        elif choice == "2":
            do_deposit()
        elif choice == "3":
            do_withdrawal()
        elif choice == "4":
            view_transaction_history()
        elif choice == "5":
            print("Thank you for using SecureSave. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

run_Securesave_app()
```
